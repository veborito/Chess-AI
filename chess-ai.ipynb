```python
import random

player = 0
tour = 1
while True:
    print(f"Tour : {tour}\n")
    if player == 0:
        player = 1
        legal_moves = str(board.legal_moves).split('(')[1][:-2].replace(" ", "").split(',')
        #no seed
        random_choice = random.randint(0, len(legal_moves) - 1)
        move = legal_moves[random_choice]
        board.push_san(move)
        print(f"AI plays {move}\n")
        print(board, "\n")
        if board.is_checkmate():
            print("well done white won !")
            break
        elif board.is_stalemate():
            print("It's draw")
            break
    if player == 1:
        player = 0
        legal_moves = str(board.legal_moves).split('(')[1][:-2].replace(" ", "").split(',')
        #no seed
        random_choice = random.randint(0, len(legal_moves) - 1)
        move = legal_moves[random_choice]
        board.push_san(move)
        print(f"AI plays {move}\n")
        print(board, "\n")
        if board.is_checkmate():
            print("well done black won !")
            break
        elif board.is_stalemate():
            print("It's draw")
            break
    tour += 1
            
        
```
